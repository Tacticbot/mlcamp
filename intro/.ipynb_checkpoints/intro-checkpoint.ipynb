```python
#vector multiplication
```

```python
def vector_vector_multiplication(u, v):
    assert u.shape[0] == v.shape[0]
    n = v.shape[0]

    result = 0.0

    for i in range(n):
        result += u[i] * v[i]
    return result
    
```

```python
u = np.array([2, 3, 4])
v = np.array([5, 6, 7])

vector_vector_multiplication(u, v)
```

```python
u.dot(v)
```

```python
#matrix vector multiplication
```

```python
def matrix_vector_multiplication(u, v):
    assert u.shape[1] == v.shape[0]
    n = u.shape[0]
    k = v.shape[0]

    result = np.zeros(n)

    for i in range(n):
        for j in range(k):
            result[i] += u[i,j] * v[j]]
    return result
    
```

```python
u = np.array([
    [2,3,4],
    [1,2,1],
    [3,4,5]
])
v = np.array([5, 6, 7])
```

```python
matrix_vector_multiplication(u, v)
```

```python
np.dot(u,v)
```

```python
def matrix_matrix_multiplication(u, v):
    assert u.shape[1] == v.shape[0]
    n = u.shape[0]
    k = v.shape[1]

    result = np.zeros([n, k])

    for i in range(n):
        for j in range(k):
            result[i, j] = np.sum(u[i,:] * v[:,j])
    return result
    
```

```python
u = np.array([
    [2,3,4],
    [1,2,1],
    [3,4,5]
])
v = np.array([
    [5, 6, 7, 4],
    [5, 8, 2, 1],
    [3, 2, 3, 2]
])
```

```python
matrix_matrix_multiplication(u, v)
```

```python
u.dot(v)
```
